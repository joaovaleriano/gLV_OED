```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
```

```python
df = pd.read_excel("Downloads/Dataset_S1.xlsx", sheet_name="MmuE", index_col=0)
df
```

```python
df.values[1:-1,:-2]
```

```python
A = df.values[1:-1,:-2].astype(np.float64)
plt.imshow(A*(1-np.eye(A.shape[0])), cmap="coolwarm", vmin=-np.abs(A).max(), vmax=np.abs(A).max())
plt.colorbar()
plt.show()
```

```python
A_cross_int = (A*(1-np.eye(A.shape[0]))).flatten()
A_cross_int = A_cross_int[A_cross_int!=0]
```

```python
plt.hist(A_cross_int, density=True)
```

```python
A_cross_int.mean()
```

```python
plt.plot(np.linspace(0, 1, len(A_cross_int)), np.sort(A_cross_int))
plt.plot(np.linspace(0, 1, 1000), np.sort(np.random.normal(A_cross_int.mean(), A_cross_int.std(), 1000)))
```

```python
plt.plot(np.sort(A_cross_int), (np.arange(len(A_cross_int))+1)/len(A_cross_int))
plt.plot(np.sort(A_cross_int), st.norm.cdf(np.sort(A_cross_int), A_cross_int.mean(), A_cross_int.std()))
np.abs((np.arange(len(A_cross_int))+1)/len(A_cross_int)-st.norm.cdf(np.sort(A_cross_int), A_cross_int.mean(), A_cross_int.std())).max()
```

```python
plt.plot(np.sort(A_cross_int), (np.arange(len(A_cross_int))+1)/len(A_cross_int))
plt.plot(np.sort(A_cross_int), st.laplace.cdf(np.sort(A_cross_int), A_cross_int.mean(), A_cross_int.std()/2**0.5))
np.abs((np.arange(len(A_cross_int))+1)/len(A_cross_int)-st.laplace.cdf(np.sort(A_cross_int), A_cross_int.mean(), A_cross_int.std()/2**0.5)).max()
```

```python
plt.plot(np.sort(A_cross_int), (np.arange(len(A_cross_int))+1)/len(A_cross_int))
plt.plot(np.sort(A_cross_int), st.cauchy.cdf(np.sort(A_cross_int), A_cross_int.mean(), A_cross_int.std()/2))
np.abs((np.arange(len(A_cross_int))+1)/len(A_cross_int)-st.cauchy.cdf(np.sort(A_cross_int), A_cross_int.mean(), A_cross_int.std()/2)).max()
```

```python
x = np.linspace(-2, 2, 1000)

plt.hist(A_cross_int, 30, histtype="step", density=True)
plt.plot(x, st.norm.pdf(x, A_cross_int.mean(), A_cross_int.std()))
plt.plot(x, st.laplace.pdf(x, A_cross_int.mean(), A_cross_int.std()/2**0.5))
plt.plot(x, st.cauchy.pdf(x, A_cross_int.mean(), A_cross_int.std()/1))
# plt.yscale("log")
plt.show()
```

```python
plt.hist(A_cross_int, 30, histtype="step", density=True)
plt.hist(st.laplace.rvs(A_cross_int.mean(), A_cross_int.std()/2**0.5, 110), 30, histtype="step", density=True)
plt.yscale("log")
plt.show()
```

```python
plt.hist(A.flatten(), 30, histtype="step", density=True)
plt.hist(st.laplace.rvs(A.flatten().mean(), A.flatten().std()/2**0.5, 110), 30, histtype="step", density=True)
plt.yscale("log")
plt.show()
```
